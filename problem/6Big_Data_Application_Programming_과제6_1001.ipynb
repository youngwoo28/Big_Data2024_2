```python
import random

def conduct_election(votes, candidates):
    # 각 후보자에 대한 득표수를 저장할 리스트 초기화
    results = [0] * len(candidates)

    # 투표 진행
    for i in range(1, votes + 1):
        # 무작위로 후보자 선택
        vote = random.randint(0, len(candidates) - 1)
        results[vote] += 1

        # 투표 진행률 계산
        progress_percentage = (i / votes) * 100
        print()
        # 현재 투표 상황 출력
        print(f"[투표 진행률]: {progress_percentage:.2f}%, {i}명 투표 => {candidates[vote]}")

        # 각 후보자의 득표율 및 득표수 출력
        print_candidate_status(candidates, results, i, votes)

    return results

def print_candidate_status(candidates, results, current_vote_count, total_votes):
    for i, (candidate, result) in enumerate(zip(candidates, results), start=1):
        vote_percentage = (result / total_votes) * 100
        print(f"[기호:{i}] {candidate}: {vote_percentage:6.2f}%  (투표수: {result})")
    print()

def prevent_tie(results, candidates):
    """동률 발생을 방지하는 함수"""
    max_votes = max(results)
    if results.count(max_votes) > 1:  # 동률 발생 시
        min_index = results.index(max(results))  # 최소 득표자를 찾아 마지막 표를 그에게 주기
        results[min_index] += 1
        print(f"[동률 발생 방지] 마지막 표는 {candidates[min_index]} 후보에게 추가되었습니다.")

def print_results(candidates, results):
    # 최대 득표수 찾기
    max_votes = max(results)
    winner_index = results.index(max_votes)

    print(f"[투표결과] 당선인: {candidates[winner_index]} ")

def main():
    # 1. 총 투표수 입력
    votes = int(input("총 진행할 투표수를 입력해 주세요."))
    while votes < 1 or votes > 10000:
        print("투표수는 1 ~ 10000 사이의 값이어야 합니다.")
        votes = int(input("총 진행할 투표수를 다시 입력해 주세요 (1 ~ 10000): "))

    # 2. 후보자 수 입력
    count = int(input("가상 선거를 진행할 후보자 인원을 입력해 주세요."))
    while count < 2 or count > 10:
        print("후보자 인원은 2 ~ 10 사이의 값이어야 합니다.")
        count = int(input("후보자 인원을 다시 입력해 주세요 (2 ~ 10): "))

    # 3. 후보자 이름 입력
    candidates = []
    for i in range(1, count + 1):
        candidate_name = input(f"{i}번째 후보자 이름을 입력해 주세요.: ")
        while len(candidate_name) >= 10:
            print("후보자 이름은 10자 미만이어야 합니다.")
            candidate_name = input(f"{i}번째 후보자 이름을 다시 입력해 주세요 (10자 미만): ")
        candidates.append(candidate_name)

    # 4. 투표 진행 및 결과 수집
    print()
    results = conduct_election(votes, candidates)

    # 5. 결과 출력
    print_results(candidates, results)

if __name__ == "__main__":
    main()
```
